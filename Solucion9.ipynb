```python
import pandas as pd

tbl0 = pd.read_csv("tbl0.tsv", sep="\t")
tbl1 = pd.read_csv("tbl1.tsv", sep="\t")
tbl2 = pd.read_csv("tbl2.tsv", sep="\t")


_c0=[ltr for ltr in tbl0._c1]

_c1=[]
data=[]
_c0= sorted(set(_c0))
val=''

for j in _c0:
    for i in sorted(tbl0._c2[tbl0['_c1'].str.contains(j)]):
        if val=='':
            val=str(i)
        else:
            val=str(val)+':'+str(i)
    data.append((j, val))
    val=''

mi_df= pd.DataFrame(data=(data), columns=('_c0','_c1'))
result = pd.pivot_table(mi_df, values='_c1', index=['_c0','_c1'])
  
result.reset_index('_c1')

```
